Data Pre-Processing

```python
print(parkinsons_data.columns)
```

```python
X = parkinsons_data.drop(columns=[ 'status'], axis=1)
Y = parkinsons_data['status']
```

```python
# samples with stratify=Y is the standard practice for high accuracy.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2, stratify=Y)
print("X_train shape: ", X_train.shape)
print("Y_train shape: ", Y_train.shape)
print("X_test shape: ", X_test.shape)
print("Y_test shape: ", Y_test.shape)
```

```python
scaler = MinMaxScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Model Training


```python
# Setup the SVM with poly kernel
poly_svm = SVC(kernel='poly', probability=True, class_weight='balanced', random_state=2)
```

```python
# 5. Hyperparameter Tuning for the SVM
param_grid_poly = {
    'C': [0.1, 1, 10, 100],        # Regularization (how strict the boundary is)
    'degree': [2, 1],              # How many times the curve can bend
    'gamma': ['scale', 'auto'],    # Curve influence
    'coef0': [0.0, 1.0]            # Independent term that helps shift the curve
}
```

```python
print("\nTuning SVM parameters...")
grid = GridSearchCV(poly_svm, param_grid_poly, cv=5, scoring='accuracy', n_jobs=-1)
grid.fit(X_train_scaled, Y_train)
```

```python
# Lock in the best model
best_svm_model = grid.best_estimator_
print("Best Parameters:", grid.best_params_)
```

```python
X_test_prediction = best_svm_model.predict(X_test_scaled)
print("\nFinal Accuracy:", accuracy_score(Y_test, X_test_prediction))
```

```python
print("\nConfusion Matrix:")
print(confusion_matrix(Y_test, X_test_prediction))
```

```python
print("\nClassification Report:")
print(classification_report(Y_test, X_test_prediction))
```

```python
def predict_parkinsons(input_data):
    # 1. Change input data to a numpy array
    input_data_as_numpy_array = np.asarray(input_data)

    # 2. Reshape the numpy array as we are predicting for one instance
    input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

    # 3. Create a DataFrame with the specific column names
    # This tells the scaler exactly which feature corresponds to which number
    input_df = pd.DataFrame(input_data_reshaped, columns=X.columns)

    # 4. STANDARDIZE using the DataFrame 
    std_data = scaler.transform(input_df)

    # 5. Make prediction
    prediction = best_svm_model.predict(std_data)

    if (prediction[0] == 0):
      print("Result: The Person is HEALTHY")
    else:
      print("Result: The Person has PARKINSON'S")


input_data = (116.676, 137.871, 111.366, 0.00997, 0.00009, 0.00502, 0.00698, 0.01505, 0.05492, 0.517, 0.02924, 0.04005, 0.03772, 0.08771, 0.01353, 20.644, 0.434969, 0.819235, -4.117501, 0.334147, 2.405554, 0.368975)
predict_parkinsons(input_data)
```
