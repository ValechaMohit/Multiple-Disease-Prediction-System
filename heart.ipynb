Splitting the Features and Target


```python
# --- 2. Data Preprocessing ---
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
```

```python
cols_to_scale = ['trestbps', 'chol', 'thalach', 'oldpeak']
```

```python
# Apply the scaler exclusively to those columns. 
# 'age', 'sex', 'cp', etc., remain in their raw, human-readable format!
X_train_scaled[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])
X_test_scaled[cols_to_scale] = scaler.transform(X_test[cols_to_scale])
```

Model Training

```python

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_scaled, Y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
# Accuracy Score

print(f"Accuracy of Data:     {accuracy_score(Y_test, y_pred):.4f}")

print("\nConfusion Matrix:")
print(confusion_matrix(Y_test, y_pred))
```

```python
# Classification Report - Gives Precision/Recall details
print("\nClassification Report:")
print(classification_report(Y_test, y_pred))
```

Predictive System

```python
input_data = (57, 0, 1, 130, 236, 0, 0, 174, 0, 0.0, 1, 1, 2)


# Convert the input tuple directly into a DataFrame
# We use 'X.columns' to ensure the column names match exactly what the model expects
input_df = pd.DataFrame([input_data], columns=X.columns)

print("Input Data Processed:")
print(input_df)



# Make prediction
prediction = model.predict(input_df)

print(f"\nModel Prediction: {prediction[0]}")

if (prediction[0] == 0):
  print('The person does not have a Heart Disease ')
else:
  print('The person has Heart Disease ')
```
